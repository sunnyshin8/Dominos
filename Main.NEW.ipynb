```python
#creating charbot using python speech recognition library and text to speech library
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime
import wikipedia
import pyjokes
import webbrowser
```

```python
#initializing the speech recognition
listener = sr.Recognizer()
engine = pyttsx3.init()
```

```python
#setting the voice
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)
#fetching the voice
def talk(text):
    engine.say(text)
    engine.runAndWait()
```

```python
#ben10 voice
def ben10(text):
    engine.setProperty('voice', voices[0].id)
    engine.say(text)
    engine.runAndWait()
```

```python
#listening the voice
def take_command():
    try:
        with sr.Microphone() as source:
            print('listening...')
            voice = listener.listen(source)
            command = listener.recognize_google(voice)
            command = command.lower()
            if 'charbot' in command:
                command = command.replace('charbot', '')
                print(command)
    except:
        pass
    return command
```

```python
#responding to the voice
def run_charbot():
    command = take_command()
    print(command)
    if 'play' in command:
        song = command.replace('play', '')
        talk('playing' + song)
        pywhatkit.playonyt(song)
    elif 'time' in command:
        time = datetime.datetime.now().strftime('%I:%M %p')
        talk('Current time is ' + time)
    elif 'who is' in command:
        person = command.replace('who is', '')
        info = wikipedia.summary(person, 1)
        print(info)
        talk(info)
    elif 'joke' in command:
        talk(pyjokes.get_joke())
    elif 'open' in command:
        website = command.replace('open', '')
        webbrowser.open(website)
    elif 'ben10' in command:
        ben10('I am ben10')
    elif 'stop' in command:
        talk('Goodbye')
        exit()
    else:
        talk('Please say the command again.')
        print('Listening...')
        run_charbot()
```

```python
import pandas as pd
from googletrans import Translator
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

# Load the CSV file
file_path = 'Indian Language News Text Summarization\Eng_train.csv'
df = pd.read_csv(file_path)


print(df.columns)


translator = Translator()

# Function to translate a given text to Hindi
def translate_text(text):
    try:
        translation = translator.translate(text, dest='hi')
        return translation.text
    except Exception as e:
        return text  

# Apply the translation function to the 'Article' column
df['Translated_Article'] = df['Article'].apply(translate_text)


def summarize_text(text, num_sentences=3):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LsaSummarizer()
    summary = summarizer(parser.document, num_sentences)
    return " ".join([str(sentence) for sentence in summary])


df['Generated_Summary'] = df['Translated_Article'].apply(lambda x: summarize_text(x, num_sentences=3))
print(df[['id', 'Heading', 'Translated_Article', 'Generated_Summary']].head())
```
